## Descripci√≥n del conjunto de datos

El conjunto de datos "Traffic Accident Prediction" contiene informaci√≥n sobre accidentes de tr√°fico en diversas redes viales, con 841 registros y 14 caracter√≠sticas. Las caracter√≠sticas incluyen:

## Caracter√≠sticas del Dataset

-	Weather: Condiciones clim√°ticas (Clear, Rainy, Foggy, Snowy, Stormy).
-	Road_Type: Tipo de carretera (Highway, City Road, Rural Road, Mountain Road).
-	Time_of_Day: Hora del d√≠a (Morning, Afternoon, Evening, Night).
-	Traffic_Density: Densidad del tr√°fico (0: Baja, 1: Moderada, 2: Alta).
-	Speed_Limit: L√≠mite de velocidad en la carretera.
-	Number_of_Vehicles: N√∫mero de veh√≠culos involucrados (1 a 5).
-	Driver_Alcohol: Consumo de alcohol por parte del conductor (0: No, 1: S√≠).
-	Accident_Severity: Severidad del accidente (Low, Moderate, High).
-	Road_Condition: Condiciones de la carretera (Dry, Wet, Icy, Under Construction).
-	Vehicle_Type: Tipo de veh√≠culo (Car, Truck, Motorcycle, Bus).
-	Driver_Age: Edad del conductor (18 a 70 a√±os).
-	Driver_Experience: A√±os de experiencia del conductor (0 a 50 a√±os).
-	Road_Light_Condition: Condiciones de iluminaci√≥n (Daylight, Artificial Light, No Light).
-	Accident: Ocurrencia de accidente (1: S√≠, 0: No).

La fuente de este conjunto de datos es Kaggle, y puede ser consultado en el enlace que se encuentra en referencias. 

## Tama√±o del Dataset
